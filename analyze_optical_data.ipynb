```python
'''

Load necessary packages

'''

import numpy as np
from scipy.signal import resample
import xarray as xr
import glob
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
from itertools import product 
```

```python
'''

Set basic variables for analysis

'''

# set material
material = "K1"

# set path to optical data files
path = "data/"
all_files = glob.glob(path+material+"*.csv")

# get list  of samples and layer thickness in directory
layer_str = np.sort(np.unique([f.split('-')[1] for f in all_files]))
layers = np.array([layer.replace(',','.').split("mm")[0] for layer in layer_str]).astype(np.float64)
samples = np.sort(np.unique([f.split('-')[3].split('.')[0] for f in all_files]))

# get wavelength vector (all files MUST have the same vector of wavelengths!)
wavelens = np.loadtxt(all_files[0],delimiter=",",skiprows=2)[:,0]
wavelen_step = wavelens[0] - wavelens[1]
```

```python
'''

Read all optical data into data structure in memory

'''

# make dummy data structure with correct dimension and fill with NaNs
transmittance = np.empty((1,len(layers),len(samples),(len(wavelens))))
reflectance = np.empty((1,len(layers),len(samples),(len(wavelens))))
transmittance[:] = np.nan
reflectance[:] = np.nan

# make xarray dataset with proper dimensions and metadata
ds = xr.Dataset(
    {
        "transmittance": (["material", "layer", "sample", "wavelength"], transmittance),
        "reflectance": (["material", "layer", "sample", "wavelength"], reflectance),
    },
    coords={
        "material": [material],
        "layer": layer_str,
        "sample": samples,
        "wavelength": wavelens,
    },
    attrs=dict(description="Optical data for various HGMs."),
)

# read each data file and use the metadata contained in the filename to place it in the xarray dataset
for filename in all_files:
    
    # get metadata from filename
    material, layer, mode, sample = filename.split("/")[-1].split(".")[0].split("-")

    # read file with numpy, skipping first two rows, which have comments, not data
    # first column (data[:,0]) is wavelengths and second column (data[:,1]) is optical data
    data = np.loadtxt(filename,delimiter=",",skiprows=2)
            
    # load optical data into xarray DataArray
    data = xr.DataArray(data[:,1], dims=("wavelength"), coords={"wavelength":wavelens})

    # check if the filename contained "T" or "R" and use this to sort reflectance and transmittance data
    # use ds.fillna to assign data from file to the correct location in the DataSet
    if mode == "T":
        trans_data = ds.loc[dict(material=material,layer=layer,sample=sample)].fillna(value={"transmittance":data})
        ds.loc[dict(material=material,layer=layer,sample=sample)] = trans_data
    elif mode == "R":
        refl_data = ds.loc[dict(material=material,layer=layer,sample=sample)].fillna(value={"reflectance":data})
        ds.loc[dict(material=material,layer=layer,sample=sample)] = refl_data
```

```python
'''

Plot sample-averaged R, T, and A at each thickness

'''

# choose colors
c1, c2, c3 = 'b','g','r'

# make a plot for each layer thickness
for i,layer in enumerate(layer_str):

    # make figure
    fig,ax = plt.subplots()
    
    # extract data
    trans = ds.loc[dict(material=material,layer=layer)]["transmittance"].values
    refl = ds.loc[dict(material=material,layer=layer)]["reflectance"].values

    # compute mean of optical properties across samples
    trans_mean = np.mean(trans,axis=0)
    refl_mean = np.mean(refl,axis=0)
    abs_mean = 100 - (trans_mean + refl_mean)  

    # compute standard deviations
    trans_std = np.std(trans,axis=0)
    refl_std = np.std(refl,axis=0)
    abs_std = np.sqrt(trans_std**2 + refl_std**2)
    
    # make scatter plots of each curve
    ax.scatter(wavelens, trans_mean, c=c1, s=2, label="% Transmittance")
    ax.scatter(wavelens, refl_mean, c=c2, s=2, label="% Reflectance")
    ax.scatter(wavelens[abs_mean > 0], abs_mean[abs_mean > 0], c=c3, s=2, label="% Absorptance")

    # shade 1 standard deviation from each curve
    ax.fill_between(wavelens, trans_mean-trans_std, trans_mean+trans_std,color=c1,alpha=0.25)
    ax.fill_between(wavelens, refl_mean-refl_std, refl_mean+refl_std,color=c2,alpha=0.25)
    ax.fill_between(wavelens, abs_mean-abs_std, abs_mean+abs_std,color=c3,alpha=0.25)

    # adjust plots
    ax.grid(visible=True,c='lightgray',linewidth=0.5)
    ax.set_ylim([0,100])
    ax.set_xlim([350,2500])
    ax.set_xticks(np.arange(500,2500,250))
    ax.set_yticks(np.arange(10,110,10))
    ax.set_xlabel("Wavelength (nm)")
    layer_txt = layer.replace(",",".").replace("mm"," mm")
    ax.set_title("Optical Data for "+layer_txt+" Layer of "+material+" (5 Samples)",fontsize=14)
    ax.legend(scatterpoints=5,loc="upper left")
    ax.set_axisbelow(True)
    plt.savefig('optical_data_'+material+"_"+layer+'.png', dpi=200)
```

```python
'''

Calculate mean optical properties, weighted by Arctic solar radiation spectra from Grenfell & Perovich (2008)

'''

# read radiation spectra (first column is wavelengths; second and third columns are incident solar irradiance from 2 days)
solar_spectra_file = path+"perovich_2008_data.csv"
solar_spectra = np.loadtxt(solar_spectra_file,delimiter=",",skiprows=1)[:,:2]

# resample solar spectra to data sampling rate
num_samples = len(wavelens[wavelens >= 350])
solar_spectra = resample(solar_spectra[:,1],num_samples,axis=0)

# normalize solar spectra for use weighting HGM optical properties (don't normalize the wavelengths)
norm_solar_spectra = solar_spectra/np.sum(solar_spectra)

# NOTE: the solar spectra are ordered 350...2500 while the optical data are ordered 2500..200 
# to make them consistent, we must flip the solar spectra
norm_solar_spectra = np.flip(norm_solar_spectra,axis=0)

# make array for storing optical results (each row is T,R,A at a particular layer thickness) and breakeven albedos
overall_mean_values = np.zeros((len(layers),3))
breakeven_albedos = np.zeros(len(layers))

# for each layer thickness, average optical properties across the 5 samples taken and weight by solar spectra
# NOTE: we will only consider optical data in the 350-2500 nm band since that's the range the solar spectra cover
# NOTE: all reported average values will only include wavelengths for which A > 0
for i,layer in enumerate(layer_str):
    
    # get optical data
    trans = ds.loc[dict(material=material,layer=layer)]["transmittance"].values
    refl = ds.loc[dict(material=material,layer=layer)]["reflectance"].values
    
    # extract data only from 350-2500 nm
    trans = trans[:,wavelens >= 350]
    refl = refl[:,wavelens >= 350]

    # compute mean of optical properties across samples and then compute absorptance
    trans_sample_mean = np.mean(trans,axis=0)
    refl_sample_mean = np.mean(refl,axis=0)
    abs_sample_mean = 100 - (trans_sample_mean + refl_sample_mean)  

    # compute standard deviation for optical properties
    trans_std = np.std(trans,axis=0)
    refl_std = np.std(refl,axis=0)
    abs_std = np.sqrt(trans_std**2 + refl_std**2)

    # get the indices of values for which A > 0 
    valid_idx = abs_sample_mean > 0
    
    # STEPHANIE'S METHOD
    # computed weighted average for each optical property, only for values where A > 0
    # NOTE: the weighting vector (solar spectra) must also be subset to include only wavelengths where A > 0
    trans_weighted_mean = np.average(trans_sample_mean[valid_idx],weights = norm_solar_spectra[valid_idx])
    refl_weighted_mean = np.average(refl_sample_mean[valid_idx],weights = norm_solar_spectra[valid_idx])
    abs_weighted_mean = np.average(abs_sample_mean[valid_idx],weights = norm_solar_spectra[valid_idx])

    # compute weighted average standard deviations 
    trans_weighted_std = np.average(trans_std[valid_idx],weights = norm_solar_spectra[valid_idx])
    refl_weighted_std = np.average(refl_std[valid_idx],weights = norm_solar_spectra[valid_idx])
    abs_weighted_std = np.average(abs_std[valid_idx],weights = norm_solar_spectra[valid_idx])

    # compute the break-even albedo
    r_ice = np.linspace(0,1,100)
    r = (refl_weighted_mean/100)+r_ice*(trans_weighted_mean/100)**2/(1-r_ice*(refl_weighted_mean/100))
    breakeven_albedo = r_ice[np.where(r < r_ice)[0][0]]
    
    # place values into array
    overall_mean_values[i,0] = trans_weighted_mean
    overall_mean_values[i,1] = refl_weighted_mean
    overall_mean_values[i,2] = abs_weighted_mean
    breakeven_albedos[i] = breakeven_albedo
    
    # calculate max and min possible break-even albedos based on weighted standard deviation
    uncert_breakevens = np.zeros((4))
    trans_stds = [-trans_weighted_std,trans_weighted_std]
    refl_stds = [-refl_weighted_std,refl_weighted_std]
    trans_refl_stds = list(product(trans_stds,refl_stds))
    for i, trans_refl_std in enumerate(trans_refl_stds):
        trans_uncert = trans_weighted_mean + trans_refl_std[0]
        refl_uncert = refl_weighted_mean + trans_refl_std[1]
        r = (refl_uncert/100)+r_ice*(trans_uncert/100)**2/(1-r_ice*(refl_uncert/100))
        if len(np.where(r < r_ice)[0]) > 0:
            uncert_breakevens[i] = r_ice[np.where(r < r_ice)[0][0]]
        else:
            uncert_breakevens[i] = 1
    
    # print results
    print("------------------" + layer+" of "+material+"------------------\n")
    print("--Weighted mean--           --Weighted stdev--")
    print("Transmittance: " + str(np.round(trans_weighted_mean,2)) + "             "+str(np.round(trans_weighted_std,2)))
    print("Reflectance:   " + str(np.round(refl_weighted_mean,2)) + "             "+str(np.round(refl_weighted_std,2)))
    print("Absorptance:   " + str(np.round(abs_weighted_mean,2)) + "              "+str(np.round(abs_weighted_std,2)))
    print()
    print("Break-even albedo:   " + str(np.round(breakeven_albedo,2)) + "   (min: "+str(np.round(np.min(uncert_breakevens),2))+"    max: "+str(np.round(np.max(uncert_breakevens),2))+")")
    print()
```

```python
'''

Perform linear regression, interpolate value at 0.33 mm, and plot optical properties as a function of layer thickness

'''

# perform basic linear regression
m_t,b_t = np.polyfit(layers, overall_mean_values[:,0], 1)
m_r,b_r = np.polyfit(layers, overall_mean_values[:,1], 1)
m_a,b_a = np.polyfit(layers, overall_mean_values[:,2], 1)

# interpolate to some other layer thickness
interp_layer = 0.33
t_interp = interp_layer*m_t+b_t
r_interp = interp_layer*m_r+b_r
a_interp = interp_layer*m_a+b_a

# compute the break-even albedo
r_ice = np.linspace(0,1,100)
r = (r_interp/100)+r_ice*(t_interp/100)**2/(1-r_ice*(r_interp/100))
breakeven_interp = r_ice[np.where(r < r_ice)[0][0]]

# print interpolated results
print("------------------"+str(np.round(interp_layer,2))+"mm of "+material+"------------------\n")
print("--Interpolated values--")
print("Transmittance: " + str(np.round(t_interp,2)))
print("Reflectance:   " + str(np.round(r_interp,2)))
print("Absorptance:   " + str(np.round(a_interp,2)))
print()
print("Break-even albedo:   " + str(np.round(breakeven_interp,2)))
print()

# make new figure
fig,ax = plt.subplots()

# plot overall weighted average optical data 
plt.scatter(layers,overall_mean_values[:,0],marker="o",label = "% Transmittance",c="b",s=100)
plt.scatter(layers,overall_mean_values[:,1],marker="v",label = "% Reflectance",c="g",s=100)
plt.scatter(layers,overall_mean_values[:,2],marker="s",label = "% Absorptance",c="r",s=100)

# plot best fit lines 
max_val = np.max(layers)+0.1
plt.plot(np.array([0,max_val]),np.array([0,max_val])*m_t+b_t,linestyle="--",c="b",alpha=0.5)
plt.plot(np.array([0,max_val]),np.array([0,max_val])*m_r+b_r,linestyle="--",c="g",alpha=0.5)
plt.plot(np.array([0,max_val]),np.array([0,max_val])*m_a+b_a,linestyle="--",c="r",alpha=0.5)

# fix a few plot things
ax.grid(visible=True,c='lightgray',linewidth=0.5)
ax.set_axisbelow(True)
plt.xlim(0,np.max(layers)+0.1)
plt.ylim(0,100)
plt.yticks()
plt.legend()
plt.xlabel("Layer thickness (mm)")
plt.title("Average "+material+" optical properties with layer thickness")
plt.savefig('average_optical_properties.png', dpi=200)
```
